```python
# Read the original file
df = pd.read_csv('/content/drive/MyDrive/QTL_analysis_051223/QTLARS_UCD1_2.txt', delimiter='\t', encoding='latin1')

# Get unique chromosomes in the DataFrame
unique_chromosomes = df["Chromosome"].unique()

# Iterate through each unique chromosome and create a separate file for each
for chromosome in unique_chromosomes:
    # Create a DataFrame for the specific chromosome
    chromosome_df = df[df["Chromosome"] == chromosome]

    # Define the output file name based on the chromosome
    output_file = f"qtl_data_{chromosome}.txt"

    # Save the DataFrame to a separate file
    chromosome_df.to_csv(output_file, sep='\t', index=False)

    # save the file into drive
    shutil.copy(output_file, "/content/drive/MyDrive/QTL_analysis_051223/QTL_ARS_UCD1_2")
```

```python
a=0
b=0
c=0
```

```python
# Identifying the variations
final_df = pd.DataFrame(columns=list(qtl_df.columns)+["startpos.r"]+["endpos.r"]+["startpos.q"]+["endpos.q"]+["seq.r"]+["seq.q"]+["LOV"]+["gene_type"])
for i, row in tqdm(df.iterrows()):
    # Check for the at start
    if qtl_df[(qtl_df["QTL Span1"]==row["startpos.r"]) & (qtl_df["QTL Span2"]>row["endpos.r"])].shape[0]>=1:
        tmp_df = qtl_df[(qtl_df["QTL Span1"]==row["startpos.r"]) & (qtl_df["QTL Span2"]>row["endpos.r"])].copy(deep=True)
        tmp_df["gene_type"] = "at_start"
        tmp_df["startpos.r"] = row["startpos.r"]
        tmp_df["endpos.r"] = row["endpos.r"]
        tmp_df["startpos.q"] = row["startpos.q"]
        tmp_df["endpos.q"] = row["endpos.q"]
        tmp_df["seq.r"] = row["seq.r"]
        tmp_df["seq.q"] = row["seq.q"]
        tmp_df["LOV"] = row["LOV"]
        final_df = pd.concat([final_df,tmp_df]).reset_index(drop=True)
        a=a+1
    # check for the at end
    elif qtl_df[(qtl_df["QTL Span1"]<row["startpos.r"]) & (qtl_df["QTL Span2"]==row["endpos.r"])].shape[0]>=1:
        tmp_df = qtl_df[(qtl_df["QTL Span1"]<row["startpos.r"]) & (qtl_df["QTL Span2"]==row["endpos.r"])].copy(deep=True)
        tmp_df["gene_type"] = "at_end"
        tmp_df["startpos.r"] = row["startpos.r"]
        tmp_df["endpos.r"] = row["endpos.r"]
        tmp_df["startpos.q"] = row["startpos.q"]
        tmp_df["endpos.q"] = row["endpos.q"]
        tmp_df["seq.r"] = row["seq.r"]
        tmp_df["seq.q"] = row["seq.q"]
        tmp_df["LOV"] = row["LOV"]
        final_df = pd.concat([final_df,tmp_df]).reset_index(drop=True)
        b=b+1
    # Check for the within gene
    elif qtl_df[(qtl_df["QTL Span1"]<row["startpos.r"]) & (qtl_df["QTL Span2"]>row["endpos.r"])].shape[0]>=1:
        tmp_df = qtl_df[(qtl_df["QTL Span1"]<row["startpos.r"]) & (qtl_df["QTL Span2"]>row["endpos.r"])].copy(deep=True)
        tmp_df["gene_type"] = "within_gene"
        tmp_df["startpos.r"] = row["startpos.r"]
        tmp_df["endpos.r"] = row["endpos.r"]
        tmp_df["startpos.q"] = row["startpos.q"]
        tmp_df["endpos.q"] = row["endpos.q"]
        tmp_df["seq.r"] = row["seq.r"]
        tmp_df["seq.q"] = row["seq.q"]
        tmp_df["LOV"] = row["LOV"]
        final_df = pd.concat([final_df,tmp_df]).reset_index(drop=True)
        c=c+1
    else:
        pass
```
